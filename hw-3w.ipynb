```python
with open('data-397-2019-12-26.json') as file_metro:
    stations = json.load(file_metro)
```

```python
for st in stations:
    Name= st['Name']
    if st['RepairOfEscalators']:
        Repair = f"эскалаторы на ремонте в период {st['RepairOfEscalators'][0]['RepairOfEscalators']}"
        print(f'станция: {Name} {Repair}')    
```
